# 4.5 The annulus theorem

The following is a useful result which can simplify integrating around a singularity.

```{prf:theorem} annulus theorem

If $f(z)$ is analytic on the closed annulus determined by the two simple closed paths $\mathscr{C}_1$ and $\mathscr{C}_2$ (see {numref}`annulus` below) then

$$
\oint_{\mathscr{C}_1} f(z)\mathrm{d} z = \oint_{\mathscr{C}_2} f(z)\mathrm{d} z,
$$

provided that the two paths are traversed in the same direction.
```

````{prf:proof}

This can be shown by splitting the annulus via two additional paths $\gamma_1$ and $\gamma_2$ as shown in {numref}`annulus`. Note, the paths in this figure do not have to be circular. The paths $\mathscr{C}_1$ and $\mathscr{C}_2$ are therefore split to new paths denoted $\mathscr{C}^*_1$ and $\mathscr{C}^*_2$.

```{figure} ../../../images/AnnulusTheorem.png
---
height: 250px
name: annulus
---
An annulus formed from two circular paths $\mathscr{C}_1$ and $\mathscr{C}_2$ cut via additional paths $\gamma_1$ and $\gamma_2$ separated by a small distance $\epsilon\to 0$ to leave new paths $\mathscr{C}^*_1$ and $\mathscr{C}^*_2$.
```

Now, considering the integral around the entire shape as indicated by the arrows:

$$
I &=  \int_{\mathscr{C}^*_1}f(z)\mathrm{d} z + \int_{\gamma_1} f(z)\mathrm{d} z - \int_{\mathscr{C}^*_2}f(z)\mathrm{d} z + \int_{\gamma_2}f(z)\mathrm{d} z\\
&= 0\quad \text{due to Cauchy's theorem}
$$

Note that the minus sign in the integral around $\mathscr{C}^*_2$ is because the integral is taken in the opposite direction to $\mathscr{C}^*_2$ indicated in the figure.

In the limit $\epsilon\to 0$, $\gamma_2$ is the opposite path to $\gamma_1$ (so that $\gamma_1\to -\gamma_2$):

$$
	\int_{\gamma_1} f(z)\mathrm{d} z=-\int_{\gamma_2} f(z)\mathrm{d} z\
$$

so that those two integrals cancel in $I$. Also, in this limit, the $\mathscr{C}^*_1\to \mathscr{C}_1$ and $\mathscr{C}^*_2\to \mathscr{C}_2$ and

$$
\int_{\mathscr{C}^*_1}f(z)\mathrm{d} z = \oint_{\mathscr{C}_1}f(z)\mathrm{d} z,\ \int_{\mathscr{C}^*_2}f(z)\mathrm{d} z = \oint_{\mathscr{C}_2}f(z)\mathrm{d} z,\ 
$$

then

$$
I = \oint_{\mathscr{C}_1}f(z)\mathrm{d} z-\oint_{\mathscr{C}_2}f(z)\mathrm{d} z=0
$$

which gives the required result for any simple closed paths $\mathscr{C}_1$ and $\mathscr{C}_2$.
````

```{note}
This theorem implies that, instead of integrating a function with a single singularity around a complicated closed path (avoiding the singularity), one can integrate the function around a circle with radius chosen such that the circle lies wholly within that path. This can greatly simplify the calculation.
```

````{prf:example} 
:label: example54

Evaluate $\displaystyle I=\oint_\mathscr{C} \frac{1}{z-i} \mathrm{d} z$ where the path $\mathscr{C}$ is a square centred on $z=0$ with side 4.

```{toggle}
The given function $f(z)=\frac{1}{z-i}$ has a pole at $z=i$ which lies within $\mathscr{C}$. The integration can be simplified by replacing $\mathscr{C}$ with a circle centred at $z=i$ of radius $r<1$; *i.e.*:

$$
	\mathscr{C}^*: |z-i| = r.
$$

Then, using the annulus theorem

$$
I=\oint_\mathscr{C} \frac{1}{z-i} \mathrm{d} z=\oint_{\mathscr{C}^*} \frac{1}{z-i} \mathrm{d} z=\oint_{|z-i|=r} \frac{1}{z-i} \mathrm{d} z
$$

Now, in a previous example in these lectures we showed that integrating $z^{-n}$ around the unit circle gives:

$$
	\oint z^{-n} \mathrm{d} z =
\begin{cases} & \int_0^{2\pi}e^{-in\theta}\ ie^{i\theta}\mathrm{d}\theta = \int_0^{2\pi} i \mathrm{d} \theta = 2i\pi \ \mbox{if}\ n=1\\
						 & \int_0^{2\pi} e^{-in\theta}\ ie^{i\theta}\mathrm{d}\theta = i \int_0^{2\pi} e^{i(1-n)\theta} \mathrm{d} \theta = \frac{1}{(1-n)}\left( e^{2i\pi(1-n)}-1\right)=0\ \mbox{if}\ n\neq 1,
	\end{cases}
$$

and the same is true for the more general case $f(z)=\frac{1}{z-a}$. (**prove this to yourself.**) 

So, we can read off the answer:

$$
	I=\oint_\mathscr{C} \frac{1}{z-i} \mathrm{d} z=\oint_{|z-i|=r} \frac{1}{z-i} \mathrm{d} z=2i\pi.
$$
```

````

## The multiple annulus theorem

By an analogous argument, it is simple to show that the annulus theorem can be extended to a multiple annulus.

If $\mathscr{C}_1,\ \mathscr{C}_2,\ \mathscr{C}_3,\ldots,\mathscr{C}_n,\ldots \mathscr{C}$ are $n+1$ simple closed paths such that:


1. $\mathscr{C}_i$, $i=1,\ldots,n$ lie within $\mathscr{C}$;
2. Each $\mathscr{C}_i$ lie in the exterior of $\mathscr{C}_j$ for $i\neq j$; that is, the interior simple closed paths $\mathscr{C}_i$ do not intersect.

The group of paths $\mathscr{C}_i, \mathscr{C}$ ($i=1,\ldots,n$) form a closed multiple annulus.

```{figure} ../../../images/MultipleAnnulus.png
---
height: 250px
---
A multiple annulus enclosing singularities at $z=-1$ and $z=2i$.
```

```{prf:theorem} multiple annulus theorem

If $f(z)$ is analytic on a closed multiple annulus determined by the closed paths $\mathscr{C}_1,\ \mathscr{C}_2,\ \mathscr{C}_3,\ldots,\mathscr{C}_n,\ \mathscr{C}$, then

$$
\oint_{\mathscr{C}}f(z)\mathrm{d} z = \oint_{\mathscr{C}_1}f(z)\mathrm{d} z + \oint_{\mathscr{C}_2}f(z)\mathrm{d} z + \ldots + \oint_{\mathscr{C}_n}f(z)\mathrm{d} z,
$$

provided that the $n$ paths are traversed in the same direction. 
```

This theorem is useful in the same way as above; *i.e.* if $f$ has multiple singularities within $\mathscr{C}$.

```{prf:example}
:label: example55

Consider the integral:

$$
	\oint_\mathscr{C}\frac{3}{z+1}+\frac{4}{z-2i} \mathrm{d} z
$$

where the path $\mathscr{C}$ is the circle $|z|=4$ traversed in an anticlockwise direction.

Clearly the integrand fails to be analytic at $z=-1$ and $z=2i$. Each singularity may be isolated by choosing two interior paths $\mathscr{C}_1$, $\mathscr{C}_2$ such that:

$$
	\mathscr{C}_1:\ |z+1|=1\ \mbox{and}\  \mathscr{C}_2:\ |z-2i|=1\,
$$

as illustrated in the above Figure. Then, on the multiple closed annulus, the integrand is analytic since the singularities have been removed and the theorem gives:

$$
	\oint_{\mathscr{C}} \frac{3}{z+1}+\frac{4}{z-2i} \mathrm{d} z&= \oint_{\mathscr{C}_1}\frac{3}{z+1}+\frac{4}{z-2i} \mathrm{d} z + \oint_{\mathscr{C}_2}\frac{3}{z+1}+\frac{4}{z-2i} \mathrm{d} z \\
	&= \oint_{\mathscr{C}_1}\frac{3}{z+1} \mathrm{d} z + \oint_{\mathscr{C}_2}\frac{3}{z+1} \mathrm{d}  +\oint_{\mathscr{C}_1}\frac{4}{z+2i} \mathrm{d} z + \oint_{\mathscr{C}_2}\frac{4}{z-2i} \mathrm{d}  z.
$$

Now, $\frac{1}{z-2i}$ is analytic within and on $\mathscr{C}_1$ and $\frac{3}{z+1}$ is analytic within and on $\mathscr{C}_2$ so by Cauchy's theorem, the second and third integrals are zero. The integral reduces to:

$$
	\oint_{\mathscr{C}} \frac{3}{z+1}+\frac{4}{z-2i} \mathrm{d} z
	= \oint_{\mathscr{C}_1}\frac{3}{z+1} \mathrm{d} z + \oint_{\mathscr{C}_2}\frac{4}{z-2i} \mathrm{d}  z.
$$

Employing the results we derived earlier:

$$
	\oint_\mathscr{C} (z-a)^{-n} \mathrm{d} z =
	\begin{cases} & 2i\pi \ \mbox{if}\ n=1\\
						 & 0\ \mbox{if}\ n\neq 1,
	\end{cases}
$$

when integrated around a unit circle centred on $a$, we can evaluate the above integral as

$$
	\oint_{\mathscr{C}_1}\frac{3}{z+1} \mathrm{d} z + \oint_{\mathscr{C}_2}\frac{4}{z-2i} \mathrm{d}  z=3(2i\pi )+4(2i\pi )=14i\pi.
$$
```
