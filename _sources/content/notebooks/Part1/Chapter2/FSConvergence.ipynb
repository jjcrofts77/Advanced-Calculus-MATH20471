# 2.3 Convergence of Fourier Series


## Motivation

So far we have assumed convergence of the Fourier series (FS):

$$
f(x) \sim \frac{a_0}{2} + \sum_{n=1}^\infty a_n\cos(nx) +b_n\sin(nx).
$$

Under this assumption we found that the coefficients are given by the **Euler integrals**:

$$
a_n = \frac{1}{\pi}\int_{-\pi}^\pi f(x)\cos(nx)\mathrm{d}x \quad (n=0,1,2,\ldots);\quad b_n=\frac{1}{\pi}\int_{-\pi}^\pi f(x)\sin(nx)\mathrm{d}x \quad (n=1, 2, 3, \ldots).
$$

An important question is: for these values of $a_n, b_n$, which $\displaystyle x\in\mathbb{R}$ does the above FS converge?

Let us consider an example.

**Example 2.3.1** Compute the Fourier series of the function

$$
f(x) = \begin{cases} x-\pi,&x\in(-\pi,0);\\x+\pi, & x\in(0,\pi).\end{cases}
$$ (FuncExample231)

Such that $\displaystyle f(x+2\pi) = f(x)$ outside of this region.

````{margin}
```{note}
 - $f$ is an odd function
 - the period $T=2\pi$
 - The plot shows 3 periods
```
````
```{figure} ../../../images/Ex231.png
---
height: 300px
name: FigExample231
---
Plot of the function in {eq}`FuncExample231`
```

**Solution**
```{toggle}
$f$ is *odd* so we know that $\displaystyle a_0=a_n=0 \forall n$.

$$
b_n &=\frac{1}{\pi}\int_{-\pi}^\pi f(x)\sin(nx)\mathrm{d}x\\
&= \frac{2}{\pi}\int_0^\pi (\pi+x)\sin(nx)\mathrm{d}x\\
&= \frac{2}{\pi}\left[-\frac{1}{n}(\pi+x)\cos(nx)\right]_0^\pi + \frac{2}{n\pi}\int_0^\pi\cos(nx)\mathrm{d}x\\
&= \frac{2}{\pi}\left[-\frac{2\pi}{n}(-1)^n+\frac{\pi}{n}\right].
$$

Thus $\displaystyle b_n = \frac{2}{n}\left(1-2(-1)^n\right)$ and so the Fourier series of the function in {eq}`FuncExample231` is given by

$$
f(x)\sim \sum_{n=1}^\infty \frac{2}{n}\left(1-2(-1)^n\right) \sin(nx).
$$
```

<br>

Recalling that 

````{margin}
```{note}
See the solution of Example 2.3.1.
```
````
$$
\sum_{n=1}^\infty \frac{2}{n}(1-2(-1)^n)\sin(nx)
$$

means $\displaystyle \lim_{n\to\infty} s_n(x)$ where

$$
s_n(x) = \sum_{m=1}^n \frac{2}{m}(1-2(-1)^m)\sin(mx).
$$ (PartialSums231)

The question therefore is whether $s_n(x)$ converges for each $x$ and if it does is the limit $f(x)$?

## Fourier Series Convergence Theorem

For the above example, it does appear that except at points of discontinuity the partial sums do converge to $f(x)$. At points of discontinuity they converge to zero. A similar result is true for most functions that appear in applications. (*i.e.* convergence is observed at points of continuity.) 

To present the main result of this section we first need to discuss the concept of one-sided limits.

<br>

>**Definition 2.3.1** We say that the *right-hand limit* of $f$ at $c$ is
>
>$$
 f(c_+) = \lim_{h\rightarrow 0, h>0}f(c+h),
$$
>
>if this exists. Similarly, the *left-hand limit* of $f$ at $c$ is
>
>$$
 f(c_{-}) = \lim_{h\rightarrow 0, h>0}f(c-h),
$$
>
>if this exists.

<br>

The existence part is important. Consider, for example, the function $f(x) = \sin(1/x)$, which does not have these limits at zero.

<br>

>**Definition 2.3.2** The function $f$ is *piecewise continuous* on an interval $(a, b)$ if we can divide $(a, b)$ into a finite number of sub-intervals, on each of which $f$ is defined and continuous, and the left-hand and right-hand limits at the endpoints of each sub-interval exist.

We are now in a position to state the Convergence Theorem for Fourier series.

<br>

>**Theorem 2.3.1 (Convergence theorem)** Let $f$ be a periodic function with period $2\pi$, with $f$ and $f'$ piecewise continuous on $(-\pi, \pi)$. Then the Fourier series of $f$ at $x$ converges to the value $\displaystyle \frac{1}{2}\left[f(x_+) + f(x_-)\right]$, *i.e.*
>
>$$
\frac{1}{2}\left[f(x_+) + f(x_-)\right] = \frac{1}{2}a_0 + \sum_{n=1}^\infty \left[a_n\cos(nx) + b_n \sin(nx)\right].
$$ (ThmConv)
>
>Note that if $f$ is continuous at $x$, then $f(x_+) = f(x_-) = f(x)$ so the Fourier series converges to $f(x)$.

<br>

Note that if a function is defined on an interval of length $2\pi$, we can find the Fourier series of  its periodic extension and equation {eq}`ThmConv` will then hold on the original interval. But we have to be careful at the end points of the interval: *e.g.* if $f$ is defined on $(-\pi, \pi]$ then at $\pm\pi$ the Fourier series converges to $\displaystyle \frac{1}{2}\left[f(\pi_+) + f(\pi_-)\right]$. 

<br>

**Example 2.3.3** From Example 2.2.2 (see the previous section) we have that the FS of the square wave function (see {numref}`SquareWave` below) is given by

$$
f(x)\sim \frac{4}{\pi}\sum_{m=0}^\infty \frac{1}{2m+1}\sin((2m+1)x).
$$

Thus, by Theorem 2.3.1, we have that the FS at $x=0$ converges to

$$
\frac{1}{2}\left[f(0_+) + f(0_-)\right] = \frac{1}{2}\left[1 + (-1)\right] = 0,
$$
which is clear from the partial sum plots in {numref}`SquareWave`.

Evaluating the FS at $x=\pi/2$ we obtain

$$
1 = \frac{4}{\pi}\sum_{m=0}^\infty \frac{1}{2m+1}\sin\left(\frac{(2m+1)\pi}{2}\right) = \frac{4}{\pi}\sum_{m=0}^\infty\frac{(-1)^m}{2m+1},
$$

and hence 

$$
\frac{\pi}{4} = \sum_{m=0}^\infty\frac{(-1)^m}{2m+1},
$$

which provides a series representation of $\pi$.

## Gibbs Phenomenon

As can be seen in {numref}`PartialSums231`, at a point of discontinuity, the partial sums always overshoot the limiting values. This overshoot does not tend to zero as more terms are taken, but the width of the overshooting region does tend to zero. This is known as *Gibbs Phenomenon*.
