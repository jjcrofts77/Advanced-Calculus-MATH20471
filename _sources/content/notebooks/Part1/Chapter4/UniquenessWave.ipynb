# 4.4 Uniqueness of an IBVP for a Finite String (optional)

We consider the wave equation

$$
 \frac{\partial^2 y}{\partial t^2} = c^2\frac{\partial^2 y}{\partial x^2}, \text{ for } 0<x<L, t>0,
$$ (waveUniq)

and prove a uniqueness theorem based on energy considerations. The *kinetic energy* of the string is

$$
 \frac{1}{2}\int_0^L\rho\left(\frac{\partial y}{\partial t}\right)^2\mathrm{d}x.
$$

The *stress energy* is the product of the tension and the extension, where the extension is

$$
 \int_0^L\sqrt{1+\left(\frac{\partial y}{\partial x}\right)^2}\mathrm{d}x-L&=\int_0^L\left[1+\frac{1}{2}\left(\frac{\partial y}{\partial x}\right)^2+\cdots\right]\mathrm{d}x-L,\\
 &\approx \frac{1}{2}\int_0^L\left(\frac{\partial y}{\partial x}\right)^2\mathrm{d}x.
$$

Thus

$$
E(t):=\frac{1}{2}\int_0^L\left[\rho\left(\frac{\partial y}{\partial t}\right)^2+T\left(\frac{\partial y}{\partial x}\right)^2\right]\mathrm{d}x,
$$ (energywave)

is the energy of the string. The energy appears to depend upon time but in important cases it is in fact constant.

<br>

>**Lemma 4.4.1** If $y(x,t)$ is a solution of the wave equation {eq}`waveUniq` and satisfies the boundary conditions
>
>$$
 y(0,t)=0 \text{ and } y(L,t)=0 \text{ for } t>0,
$$ (lemma41BCs)
>
>then $E(t)$ is constant for $t\geq 0$.

>**Proof** 
>
>Leibniz's rule applied to {eq}`energywave` gives
>
>$$
E'(t) = \int_0^L\left[\rho\frac{\partial y}{\partial t}\frac{\partial^2 y}{\partial t^2}+T\frac{\partial y}{\partial x}\frac{\partial^2 y}{\partial x\partial t}\right]\mathrm{d}x,
$$
>
>and, on substituting for $\rho\partial^2y/\partial t^2$ from the wave equation {eq}`waveUniq`, we find that
>
>$$
E'(t)&=T\int_0^L\left[\frac{\partial^2 y}{\partial x^2}\frac{\partial y}{\partial t}+\frac{\partial y}{\partial x}\frac{\partial^2 y}{\partial x\partial t}\right]\mathrm{d}x\\
&=T\int_0^L\frac{\partial}{\partial x}\left(\frac{\partial y}{\partial x}\frac{\partial y}{\partial t}\right)\mathrm{d}x\\
&=T\left[\frac{\partial y}{\partial x}\frac{\partial y}{\partial t}\right]_0^L,
$$
>
>hence $E'(t)=0$ since the boundary conditions {eq}`lemma41BCs` tell us that $\partial y/\partial t(0,t)=0$ and $\partial y/\partial t(L,t)=0$ for $t\geq 0$. Thus $E(t)$ is constant. $\quad\blacksquare$

>**Theorem 4.4.2** (Uniqueness) For each pair of functions $f$ and $g$, the IBVP
>
>$$
 \frac{\partial^2 y}{\partial t^2} = c^2\frac{\partial^2 y}{\partial x^2}, \text{ for } 0<x<L, t>0,
$$
>
>$$
 y(x,0)=f(x) \text{ and } \frac{\partial y}{\partial t}(x,0) = g(x) \text{ for } 0\leq x\leq L,
$$
>
>$$
 y(0,t)=0 \text{ and } y(L,t)=0 \text{ for } t>0,
$$
>
>has at most one solution. (That a solution exists we know since we have constructed a solution with the aid of separation of variables and Fourier series.)

<br>

>**Proof**
>
>Let $y(x,t)$ and $u(x,t)$ both be solutions of the IBVP. Consider the difference $w:=y-u$ and the associated energy 
>
>$$
 E(t):=\frac{1}{2}\int_0^L\left[\rho\left(\frac{\partial w}{\partial t}\right)^2 + T\left(\frac{\partial w}{\partial x}\right)^2\right]\mathrm{d}x.
$$
>
>Note that $E(t)\geq 0$. Now $w(x,t)$ is a solution of the IBVP
>
>$$
 \frac{\partial^2 w}{\partial t^2} = c^2\frac{\partial^2 w}{\partial x^2}, \text{ for } 0<x<L, t>0,
$$
>
>$$
 w(x,0)=0 \text{ and } \frac{\partial w}{\partial t}(x,0) = 0 \text{ for } 0\leq x\leq L,  
$$ (theorem42wICs)
>
>$$
 w(0,t)=0 \text{ and } w(L,t)=0 \text{ for } t>0.
$$ (theorem42wBCs)
>
>By Lemma 4.4.1 and the boundary conditions {eq}`theorem42wBCs`, $E(t)\equiv\text{ constant}$, and, in view of the initial conditions {eq}`theorem42wICs`, $E(0)=0$. Hence $E(t)=0$ for every $t>0$. Thus $\partial w/\partial x=0$, $\partial w/\partial t=0$ and $w(x,t)$ is independent of both $x$ and $t$. Using {eq}`theorem42wBCs` again tells us that $w(x,t)\equiv 0$. Thus $y=u$ and the solution is unique. $\quad \blacksquare$
