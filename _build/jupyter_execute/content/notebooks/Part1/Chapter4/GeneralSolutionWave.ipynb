# 4.4 The General Solution of the Wave Equation

The wave equation is untypical among PDEs in that it is possible to write down all the solutions. Note that if $F:\mathbb{R}\rightarrow\mathbb{R}$ is any (twice differentiable) function, and

$$
y(x,t) = F(x-ct),
$$ (waveTravSol)

then 

$$
 \frac{\partial y}{\partial x} = F'(x-ct), \quad \frac{\partial^2 y}{\partial x^2} = F''(x-ct),
$$

and

$$
 \frac{\partial y}{\partial t} = -cF'(x-ct), \quad \frac{\partial^2 y}{\partial t^2} = c^2F''(x-ct),
$$

and so {eq}`waveTravSol` is a solution of the wave equation. Equation {eq}`waveTravSol` represents a wave of constant shape propagating in the positive $x$-direction with speed $c$.

Similarly, if $G:\mathbb{R}\rightarrow\mathbb{R}$ is any (twice differentiable) function and

$$
 y(x,t) = G(x+ct),
$$ (waveTrav2)

then $y(x,t)$ is a solution of the wave equation. Equation {eq}`waveTrav2` represents a wave of constant shape propagating in the negative $x$-direction with speed $c$.

Again, the sum

$$
 y(x,t):= F(x-ct)+G(x+ct)
$$ (waveGenSol2)

is a solution of the wave equation. It will now be shown that *every* solution of the  wave equation must be of the form {eq}`waveGenSol2`.

To verify this introduce new variables

$$
 \xi := x-ct,\quad \eta:=x+ct,
$$

and seek a solution $y(x,t) = Y(\xi,\eta)$. Then

$$
 \frac{\partial y}{\partial x} = \frac{\partial Y}{\partial \xi}+\frac{\partial Y}{\partial\eta}, \qquad \frac{\partial^2 y}{\partial x^2} = \frac{\partial^2 Y}{\partial \xi^2} +2 \frac{\partial^2 Y}{\partial\xi\partial\eta} +\frac{\partial^2 Y}{\partial\eta^2}
$$

$$
 \frac{\partial y}{\partial t} = -c\frac{\partial Y}{\partial \xi}+c\frac{\partial Y}{\partial\eta}, \qquad \frac{\partial^2 y}{\partial t^2} = c^2\frac{\partial^2 Y}{\partial \xi^2} - 2c^2 \frac{\partial^2 Y}{\partial\xi\partial\eta} +c^2\frac{\partial^2 Y}{\partial\eta^2},
$$

and substitution into the wave equation gives

$$
\frac{\partial^2 Y}{\partial \xi^2} +2 \frac{\partial^2 Y}{\partial\xi\partial\eta} +\frac{\partial^2 Y}{\partial\eta^2} = \frac{\partial^2 Y}{\partial \xi^2} - 2 \frac{\partial^2 Y}{\partial\xi\partial\eta} +\frac{\partial^2 Y}{\partial\eta^2}.
$$

Hence in the new variable the wave equation transforms to the equation

$$
\frac{\partial^2Y}{\partial\xi\partial\eta} = 0,
$$

*i.e.*

$$
\frac{\partial}{\partial\xi}\left(\frac{\partial Y}{\partial\eta}\right) = 0.
$$

Thus $\partial Y/\partial\eta$ is independent of $\xi$ and is a function of $\eta$ only, say $G'(\eta)$, *i.e.*

$$
 \frac{\partial Y}{\partial \eta} = G'(\eta),
$$

and so 

$$
 \frac{\partial}{\partial\eta}[Y-G(\eta)] = 0.
$$

Thus $Y-G(\eta)$ is a function of $\xi$ only, say $F(\xi)$, and therefore

$$
 Y-G(\eta) = F(\xi),
$$

and 

$$
Y(\xi,\eta) = F(\xi)+G(\eta) \implies y(x,t) = F(x-ct)+G(x+ct).
$$

**Example 4.5** A string occupies $-\infty<x\leq 0$ and is fixed at $x=0$. A wave $y(x,t)=F(x-ct)$ is incident from $x<0$. Find the reflected wave.

The solution of the wave equation is 

$$
 y(x,t) = \underbrace{F(x-ct)}_\text{incident} + \underbrace{G(x+ct)}_\text{reflective},
$$

where $G$ is to be found. The condition $y(0,t)=0$ is to be satisfied for all $t$. Hence $F(-ct)+G(ct)=0$, for all $t$, and so $G(\theta)=-F(-\theta)$ for all $\theta$. Thus

$$
y(x,t) = \underbrace{F(x-ct)}_\text{incident} - \underbrace{F(-x-ct)}_\text{reflective}.
$$
