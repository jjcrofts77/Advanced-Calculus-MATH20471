# 6.3 Evaluation of real integrals

It is quite surprising that certain classes of complicated *real* integrals can be integrated by the residue theorem, as we shall see.

In this course we shall consider real integrals of the type

$$
\int_{-\infty}^\infty f(x)\mathrm{d}x.
$$

Such an integral is called an *improper integral*, and it has the meaning

$$
 \int_{-\infty}^\infty f(x)\mathrm{d}x = \lim_{a\rightarrow\infty}\int_a^0 f(x)\mathrm{d}x + \lim_{b\rightarrow\infty}\int_0^b f(x)\mathrm{d}x.
$$ (label1)

If both limits exist then we write

$$
 \int_{-\infty}^\infty f(x)\mathrm{d}x = \lim_{R\rightarrow\infty}\int_{-R}^R f(x)\mathrm{d}x.
$$ (label2)

The limit in {eq}`label2` is called the **Cauchy principal value** of the integral. It is written

$$
\mathrm{P} \int_{-\infty}^\infty f(x)\mathrm{d}x.
$$

It may exist even if the limits in {eq}`label1` do not. For example,

$$
\lim_{R\rightarrow\infty}\int_{-R}^R x\mathrm{d}x = \lim_{R\rightarrow\infty}\left(\frac{R^2}{2}-\frac{R^2}{2}\right) = 0, \quad\text{but}\quad \lim_{b\rightarrow\infty}\int_0^b x\mathrm{d}x = \infty.
$$

Now, we assume that the function $f(x)$ is a real rational function, *i.e.*

$$
 f(x) =\frac{P(x)}{Q(x)},
$$

whose denominator is non-zero for all real $x$ and is of degree at least 2 units higher than the degree of the numerator. Note that $P$ and $Q$ are polynomials in $x$. Examples of such functions include

$$
 \frac{1}{x^2}, ~ \frac{x^2+1}{x^4+1} ~\text{ and }~ \frac{1}{(x^2-2x+5)^2}.
$$

In particular, for such functions we have that the limits in {eq}`label1` and {eq}`label2` exist.

Consider the corresponding contour integral ($f$ the same as above)

$$
\oint_\mathcal{C}f(z)\mathrm{d}z
$$

around the path $\mathcal{C}$ in Figure 8. Since $f(x)$ is rational, $f(z)$ has finitely many poles in the upper half-plane, and if we choose $R$ large enough, then $\mathcal{C}$ encloses all these poles. By the residue theorem we then obtain

$$
 \oint_\mathcal{C}f(z)\mathrm{d}z = \int_Sf(z)\mathrm{d}z + \int_{-R}^Rf(x)\mathrm{d}x = 2\pi i\sum_j \mathrm{Res}(f; z_j),
$$ (label3)

where the sum consists of all residues of $f(z)$ at the points in the upper half-plane at which $f(z)$ has a pole.

From {eq}`label3` we have 

$$
\int_{-R}^Rf(x)\mathrm{d}x = 2\pi i\sum_j \mathrm{Res}(f;z_j) - \int_Sf(z)\mathrm{d}z.
$$ (label4)

Importantly, it can be shown that (using the ML-inequality, which is outside the scope of the current course) in the $R\rightarrow\infty$ limit the integral over the semicircle $S$ approaches zero, *i.e.*

$$
 \int_Sf(z)\mathrm{d}z\rightarrow 0 \text{ as } R\rightarrow\infty, 
$$

and so in the $R\rightarrow\infty$ limit {eq}`label4` yields the result

$$
\int_{-\infty}^\infty f(x)\mathrm{d}x = 2\pi i\sum_j \mathrm{Res}(f;z_j).$$ (label5)

Here the sum consists of all residues of $f(z)$ at the points in the upper half-plane at which $f(z)$ has a pole.


````{prf:example}
:label: example69

Show that

$$
 \int_0^\infty \frac{\mathrm{d}x}{1+x^4} = \frac{\pi}{2\sqrt{2}}. 
$$

**Solution**

The function $\displaystyle f(x)=1/(1+x^4)$ is a rational function and has a denominator 4 units higher than its numerator and so we can apply the above techniques to solve the integral. 

```{figure} ../../../images/ResSing.png
---
height: 250px
---
Singularities of $\displaystyle f(z) = 1/(1+z^4)$
```

Indeed, $\displaystyle f(z)=1/(1+z^4)$ has 4 order 1 poles at the points (see the sketch)

$$
 z_1 = e^{\pi i/4},\quad z_2 = e^{3\pi i/4},\quad z_3 = e^{-3\pi i/4}, \quad z_4 = e^{-\pi i/4}.
$$

The first two poles lie in the upper half-plane. From earlier we have that

$$
\mathrm{Res}(f; z_1) &=\left[\frac{1}{\left(1+z^4\right)'}\right]_{z=z_1} = \left.\frac{1}{4z^3}\right|_{z=z_1} = \frac{1}{4}e^{-3\pi i/4} = -\frac{1}{4}e^{\pi i/4}\\
\mathrm{Res}(f; z_2) &=\left[\frac{1}{\left(1+z^4\right)'}\right]_{z=z_2} = \left.\frac{1}{4z^3}\right|_{z=z_2} = \frac{1}{4}e^{-9\pi i/4} = \frac{1}{4}e^{-\pi i/4}.
$$

Here we have used the fact that $\displaystyle e^{\pi i} = -1$ and $\displaystyle e^{-2\pi i}=1$. 

From {eq}`label5` we have 

$$
\int_{-\infty}^\infty \frac{\mathrm{d}x}{1+x^4} = -\frac{2\pi i}{4}\left(e^{\pi i/4}-e^{-\pi i/4}\right) = -\frac{2\pi i}{4}\cdot 2i\cdot\sin\left(\frac{\pi}{4}\right) = \frac{\pi}{\sqrt{2}}.
$$

Since $\displaystyle 1/(1+x^4)$ is an even function, we thus obtain, as asserted,

$$
 \int_0^\infty\frac{\mathrm{d}x}{1+x^4} = \frac{1}{2} \int_{-\infty}^\infty\frac{\mathrm{d}x}{1+x^4} = \frac{\pi}{2\sqrt{2}}.
$$
````
